```python
# Example to search for the best classifier using the ML library scikit-learn (or sklearn) https://scikit-learn.org

def multi_classifiers():
  """
  Randomly select one classifier
  """
  from sklearn.neural_network import MLPClassifier
  from sklearn.neighbors import KNeighborsClassifier
  from sklearn.svm import SVC
  from sklearn.gaussian_process import GaussianProcessClassifier
  from sklearn.gaussian_process.kernels import RBF
  from sklearn.tree import DecisionTreeClassifier
  from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
  from sklearn.naive_bayes import GaussianNB
  from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

  import random 

  names = ["Nearest Neighbors", 
           "Linear SVM", 
           "RBF SVM", 
           "Gaussian Process",
          "Decision Tree", 
           "Random Forest", 
           "Neural Net", 
           "AdaBoost",
          "Naive Bayes"]

  classifiers = [
      KNeighborsClassifier(3),
      SVC(kernel="linear", C=0.025),
      SVC(gamma=2, C=1),
      GaussianProcessClassifier(1.0 * RBF(1.0)),
      DecisionTreeClassifier(max_depth=5),
      RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
      MLPClassifier(alpha=1, max_iter=1000),
      AdaBoostClassifier(),
      GaussianNB()]

  cl_num=random.randint(0,len(classifiers)) # randomly pick one classifier from the list of classifiers
  return names[cl_num],classifiers[cl_num]  
```

```python
name_clf, clf = multi_classifiers()
```
