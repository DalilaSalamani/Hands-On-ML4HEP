```python
def weatherForcast(humidity,previousTemp,weights):
    w1=weights[0]
    w2=weights[1]
    if previousTemp<20:
        return w1*previousTemp - w2*humidity
    else :
        return w1*previousTemp + w2*humidity
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
# Plot values from the root file 

plt.figure(figsize=(10,8))
plt.plot( np.array(tree['tree']['lepton_pt']) )
plt.show()
```

```python
# Scatter plot 

plt.figure(figsize=(10,8))
plt.scatter(np.concatenate(energyL2[0]),np.concatenate(energyL2[100]) )
plt.show()
```

```python
# Histogram

plt.figure(figsize=(10,8))
plt.hist( ([ np.sum(energyL2[pId]) for  pId in range(len(energyL2)) ]) ,bins=100)
plt.xlabel("Energy [MeV] ")
plt.ylabel("Frequency of events")
plt.title("Distribution of truth energies, particle type : Gamma "   )
plt.show()
```

```python
# Image

plt.figure( figsize=(12,10) )
plt.imshow( energyL2[10] , norm=LogNorm() ) 
plt.colorbar()
plt.show()
```
